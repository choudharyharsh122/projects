```python
class Location:

  def __init__(self, x, y):
    self.x = x
    self.y = y

```

```python
locations = []

locations.append(Location(-1,-2))
locations.append(Location(2,3))
locations.append(Location(-1,-4))
locations.append(Location(5,5))
locations.append(Location(1,2))
locations.append(Location(3,1))
locations.append(Location(2,-2))
locations.append(Location(-1,2))
locations.append(Location(2,4))
locations.append(Location(3,-2))
locations.append(Location(2,-3))
locations.append(Location(3,3))
locations.append(Location(1,1))
locations.append(Location(-3,2))
locations.append(Location(-3,-2))
locations.append(Location(-3,3))

centroids = [Location(0,0), Location(0,1), Location(1,0)]
```

```python
def find_and_set_min(location, centroids, cen_loc_list):
  min_list = []
  min_list.append((location.x - centroids[0].x)**2 + (location.y - centroids[0].y)**2)
  min_list.append((location.x - centroids[1].x)**2 + (location.y - centroids[1].y)**2)
  min_list.append((location.x - centroids[2].x)**2 + (location.y - centroids[2].y)**2)
  min_index = min_list.index(min(min_list))
  cen_loc_list[min_index].append(location)
```

```python
def reshuffle_centroids(cen_loc_list, centroids):
  for i in range(3):
    sum_x = 0
    sum_y = 0
    for j in range(len(cen_loc_list[i])):
      sum_x = sum_x + cen_loc_list[i][j].x
      sum_y = sum_y + cen_loc_list[i][j].y
    if len(cen_loc_list[i]) != 0:
      new_cent_x  = sum_x/len(cen_loc_list[i])
      new_cent_y  = sum_y/len(cen_loc_list[i])
      centroids[i].x = new_cent_x
      centroids[i].y = new_cent_y
```

```python
iterations = 0
cen_loc_list = []
cen_loc_list.append([])
cen_loc_list.append([])
cen_loc_list.append([])
while iterations < 50:
  cen_loc_list = []
  cen_loc_list.append([])
  cen_loc_list.append([])
  cen_loc_list.append([])
  for i in range(len(locations)):
    find_and_set_min(locations[i], centroids, cen_loc_list)
    reshuffle_centroids(cen_loc_list, centroids)
  iterations = iterations + 1

for k in range(len(cen_loc_list[0])):
  print("(",cen_loc_list[0][k].x, ",",cen_loc_list[0][k].y,")")
print("\n")
print("Centroid: (",centroids[0].x,",",centroids[0].y,")")
```
